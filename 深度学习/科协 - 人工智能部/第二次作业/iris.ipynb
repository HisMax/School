首先，假设我们有高中的基本统计知识

可以想到的就是将数据用图表简单地出来

看看三种鸢尾花的数据有什么直观上的特征与不同。

---

````python
import matplotlib.pyplot as plt
mo = data.data.reshape(600)
a = ['花萼长度','花萼宽度','花瓣长度','花瓣宽度']
for i in range(4):
    for j in range(4):
        if(i==j):continue
        plt.scatter(mo[i:600:4],mo[j:600:4],color='#ef475d',lw=2)
        plt.xlabel(a[i])
        plt.ylabel(a[j])
        plt.title(a[i]+'与'+a[j]+'的关系')
        plt.show()
````
